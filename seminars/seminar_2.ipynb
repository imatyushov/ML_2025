# Категориальные признаки

Загружаем датасет

https://scikit-learn.org/stable/auto_examples/preprocessing/plot_target_encoder.html#sphx-glr-auto-examples-preprocessing-plot-target-encoder-py

```python
wine_reviews = fetch_openml(data_id=42074, as_frame=True)
df = wine_reviews.frame
df.head()
```

```python
df.dropna()
```

```python
numerical_features = ["price"]
categorical_features = [
    "country",
    "province",
    "region_1",
    "region_2",
    "variety",
    "winery",
]
target_name = "points"

X = df[numerical_features + categorical_features]
y = df[target_name]
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, TargetEncoder
```

```python
for col_name in categorical_features:
    print(col_name, X[col_name].nunique())
```

```python
one_hot_encoder = OneHotEncoder(handle_unknown="ignore", max_categories=20, sparse_output=False)
target_encoder = TargetEncoder(target_type="continuous")
ordinal_encoder = OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1)
```

```python
transformed_data = one_hot_encoder.fit_transform(X[categorical_features])
encoded_data = pd.DataFrame(transformed_data, index=X.index)
encoded_data
concatenated_data = pd.concat([X, encoded_data], axis=1)
concatenated_data
```

```python
transformed_data = ordinal_encoder.fit_transform(X[categorical_features])
encoded_data = pd.DataFrame(transformed_data, index=X.index)
encoded_data
concatenated_data = pd.concat([X, encoded_data], axis=1)
concatenated_data
```

```python
transformed_data = target_encoder.fit_transform(X[categorical_features], y)
encoded_data = pd.DataFrame(transformed_data, index=X.index)
encoded_data
concatenated_data = pd.concat([X, encoded_data], axis=1)
concatenated_data
```
